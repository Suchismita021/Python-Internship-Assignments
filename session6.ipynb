```python
#1.Write a python program using inbuilt function to Remove all consonants from each string

words = ["python", "Internship", "suchi"]

vowels = "aeiouAEIOU"

for word in words:
    result = ""

    for ch in word:
        if ch in vowels:
            result = result + ch

    print(result)
```

```python
#6. Write a python program using inbuilt function to Double the word at every index divisible by 4

words = ["one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]

for i in range(len(words)):

    if i % 4 == 0:
        words[i] = words[i] * 2

print(words)
```

```python
#2. Write a python program using inbuilt function to Filter strings that alternate vowels and consonants

words = ["Data", "Banana", "cat", "abcd"]

vowels = "aeiouAEIOU"

def check(word):
    for i in range(len(word) - 1):
        if (word[i] in vowels) == (word[i + 1] in vowels):
            return False
    return True

result = list(filter(check, words))

print(result)
```

```python
#3.⁠ ⁠Write a python program using inbuilt function to Keep strings that contain at least 3 different vowels

words = ["Education", "Apple", "Idea", "Python"]
vowels = "aeiouAEIOU"

for word in words:
    found = ""
    for ch in vowels:
        if ch in word:
            found = found + ch
    if len(found) >= 3:
        print(word)
```

```python
#5.⁠ ⁠Write a python program using inbuilt function to Reduce list to a string of only digits extracted from all words

from functools import reduce
words = ["ab@c12", "python45", "hello7"]

def get_digits(result, word):

    for ch in word:
        if ch.isdigit():
            result = result + ch

    return result

digits = reduce(get_digits, words, "")

print(digits)
```

```python
#8. ⁠Write a python program using inbuilt function to Return list of (index, word) for words that are palindromes

words = ["madam", "python", "level", "hello", "noon"]

for i in range(len(words)):

    if words[i] == words[i][::-1]:
        print(i, words[i])
```
