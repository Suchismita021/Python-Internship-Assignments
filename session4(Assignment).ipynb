```python
#5. Write a function to find largest of 3 numbers.

def largest():
    a = int(input("Enter 1st no.: "))
    b = int(input("Enter 2nd no.: "))
    c = int(input("Enter 3rd no.: "))
   
    if a>b and a>c:
        print(a)
    elif b>c:
        print(b)
    else:
        print(c)

largest()
```

```python
#6. write a function to count number of vowels in string

def vowels():
    s = input("Enter a string: ")
    count = 0

    for i in s:
        if i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u' or i == 'A' or i == 'E' or i == 'I' or i == 'O' or i == 'U':
            count = count + 1

    print("Number of vowels:", count)

vowels()
```

```python
#9. Find the maximum number in a list using reduce

from functools import reduce

def maximum(x,y):
    if x > y:
        return x
    else:
        return y

list2 = [10,25,8,50,1]

result = reduce(maximum, list2)
print(result)
```

```python
#10. Take a list of numbers and add 3 to the numbers divisible by 3 in that list.

def add_three(x):
    if x % 3 == 0:
        return x + 3
    else:
        return x

list1 = [3, 5, 6, 8, 9, 12]

result = list(map(add_three, list1))

print(result)
```

```python
#11. Take a list of numbers, even and odd both, and then find the product of only even numbers.

from functools import reduce

def even(x):
    return x % 2 == 0

def multiply(x, y):
    return x * y

list1 = [2, 3, 4, 5, 6, 7, 8]

even_numbers = list(filter(even, list1))

result = reduce(multiply, even_numbers)

print(result)
```

```python
#12. Take a list of numbers, filter out the odd numbers, then double the odd numbers, after which reduce them to a single value by subtracting each number.

from functools import reduce

l1 = [2, 3, 4, 5, 6, 7, 9, 11]

odd = list(filter(lambda x: x % 2 != 0, l1))
print("Odd numbers:", odd)

double = list(map(lambda x: x * 2, odd))
print("Doubled:", double)

result = reduce(lambda x, y: x - y, double)
print("Result:", result)
```

```python
#13. Take a list of numbers, find the numbers divisible by 5 and return their average.

l1 = [5, 8, 10, 12, 15, 20, 25]

div5 = list(filter(lambda x: x % 5 == 0, l1))
print("Divisible by 5:", div5)

average = sum(div5) / len(div5)

print("Average:", average)
```
